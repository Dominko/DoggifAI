Removes the debug lines from the sample output files generated durign training

```python
file = "../../datasets/outputs/small_ft/val/samples_output"
output = "../../datasets/outputs/small_ft/val/processed"
with open(file, "r") as handle:
    with open(output, "w") as out_handle:
        lines = handle.readlines()
        for i in range (0, len(lines), 5):
            original = lines[i]
            generated = lines[i+2]
            out_handle.write(original)
            out_handle.write(generated + "\r\n")
```

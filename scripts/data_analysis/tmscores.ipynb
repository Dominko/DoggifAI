```python
import tmscoring
import os

worst_natural = '../../datasets/alphafold_output_clean/worst_aligned_natural.pdb'
worst_de_novo = '../../datasets/alphafold_output_clean/worst_aligned_de_novo.pdb'
median_natural = '../../datasets/alphafold_output_clean/median_aligned_natural.pdb'
median_de_novo = '../../datasets/alphafold_output_clean/median_aligned_de_novo.pdb'

alignment = tmscoring.TMscoring(worst_natural, worst_de_novo)
alignment.optimise()
tm = alignment.tmscore(**alignment.get_current_values())
rmsd = alignment.rmsd(**alignment.get_current_values())
print(f'Worst natural vs de novo: TM-score = {tm:.4f}, RMSD = {rmsd:.4f}')

alignment = tmscoring.TMscoring(median_natural, median_de_novo)
alignment.optimise()
tm = alignment.tmscore(**alignment.get_current_values())
rmsd = alignment.rmsd(**alignment.get_current_values())
print(f'Median natural vs de novo: TM-score = {tm:.4f}, RMSD = {rmsd:.4f}')
```

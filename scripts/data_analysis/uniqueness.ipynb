```python
# Read training, validation and test data from the given files
train_input_file = '../../datasets/doggy_data/train_ft_output'
val_input_file = '../../datasets/doggy_data/val_ft_output'
test_input_file = '../../datasets/doggy_data/test_ft_output'

# Read training data
data = []
with open(train_input_file, 'r') as f:
    for line in f:
        # skip the first line
        if 'text' in line:
            continue
        data.append(line.strip())
with open(val_input_file, 'r') as f:
    for line in f:
        if 'text' in line:
            continue
        data.append(line.strip())
with open(test_input_file, 'r') as f:
    for line in f:
        if 'text' in line:
            continue
        data.append(line.strip())

print(len(data))
print(len(set(data)))
```

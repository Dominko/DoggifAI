# Dirty

```python
from Bio import SeqIO, SeqRecord
import os, contextlib
import pandas as pd
from tqdm import tqdm
import re

raw_data_path = "../datasets/split_heavy_variable_dog_unique_functional_0.csv"
input_file_path = "../datasets/doggy_data/train_input"
output_file_path = "../datasets/doggy_data/train_output"

mask_token = " "

df = pd.read_csv(raw_data_path, index_col=0)

# print(df.head())

cdr = df[["H-CDR1", "H-CDR2", "H-CDR3"]]

input_data = re.sub(' +', mask_token,cdr.to_string(header=False, index=False))
output_data = df.to_string(header=False, index=False).replace(" ", "")

print(output_data)

with open(input_file_path, "w") as input_file:
    input_file.write(input_data)

with open(output_file_path, "w") as output_file:
    output_file.write(output_data)
```

# Fasta -> raw

```python
from os import listdir
from os.path import isfile, join
from Bio import SeqIO

full_out_path = "../datasets/raw/All_variable_dog_unique_functional"

raw_path_1 = "../datasets/raw/Heavy_variable_dog_unique_functional.fasta"
out_path_1 = "../datasets/raw/Heavy_variable_dog_unique_functional"

raw_path_2 = "../datasets/raw/Light_variable_kappa_dog_unique_functional.fasta"
out_path_2 = "../datasets/raw/Light_variable_kappa_dog_unique_functional"

raw_path_3 = "../datasets/raw/Light_variable_lambda_dog_unique_functional.fasta"
out_path_3 = "../datasets/raw/Light_variable_lambda_dog_unique_functional"

# files = [join(raw_path, f) for f in listdir(raw_path) if isfile(join(raw_path, f))]

with open(full_out_path, "w") as f:
    with open(raw_path_1) as handle:
        with open(out_path_1, "w") as f2:
            for record in SeqIO.parse(handle, "fasta"):
                f.write(str(record.seq) + "\r\n")
                f2.write(str(record.seq) + "\r\n")
    with open(raw_path_2) as handle:
        with open(out_path_2, "w") as f2:
            for record in SeqIO.parse(handle, "fasta"):
                f.write(str(record.seq) + "\r\n")
                f2.write(str(record.seq) + "\r\n")
    with open(raw_path_3) as handle:
        with open(out_path_3, "w") as f2:
            for record in SeqIO.parse(handle, "fasta"):
                f.write(str(record.seq) + "\r\n")
                f2.write(str(record.seq) + "\r\n")
```
